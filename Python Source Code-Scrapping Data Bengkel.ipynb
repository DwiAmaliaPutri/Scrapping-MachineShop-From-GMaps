```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url="https://www.google.com/maps/search/bengkel+motor+di+surabaya/@-7.2828429,112.6876011,11.55z?entry=ttu"
driver = webdriver.Chrome()
driver.get(url)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
Nama=[]
Alamat=[]
No_Telp=[]
```

```python
link=[]
items=soup.findAll('a', attrs={'class':'hfpxzc'})
for i in items:
    links=i.get('href')
    link.append(links)
```

```python
link
```

```python
for i in link:
    driver.get(i)
    a=BeautifulSoup(driver.page_source, 'html.parser')
    
    nama=a.find("h1", attrs={'class':'DUwDvf fontHeadlineLarge'})
    Nama.append(nama.text)
    
    item=a.find_all('div', attrs={'class':'Io6YTe fontBodyMedium kR99db'})
    alamat=item[0]
    Alamat.append(alamat.text)
    
    for j in range (0,len(item)):
        if item[j].text.startswith("08"):
            no_telp=item[j].text
        elif item[j].text.startswith("(03"):
            no_telp=item[j].text
        else:
            pass
        
    No_Telp.append(no_telp)
```

```python
data_bengkel=pd.DataFrame(zip(Nama, Alamat, No_Telp), columns=['Nama Bengkel', 'Alamat', 'No Telepon'])
data_bengkel
```

```python
data_bengkel.describe()
```

```python
url='https://www.google.com/maps/place/Platina+Motor/data=!4m7!3m6!1s0x2dd7fbcff5e8e9ed:0xe90904e9e5f95d5a!8m2!3d-7.277725!4d112.751013!16s%2Fg%2F11b6zvjccn!19sChIJ7eno9c_71y0RWl355ekECek?authuser=0&hl=id&rclk=1'
driver = webdriver.Chrome()
driver.get(url)
```

```python
Nama_Pelanggan=[]
Waktu_Review=[]
Rating_Penilaian=[]
Komentar=[]
```

```python
soup3 = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
for i in soup3:
    nama_bengkel = i.find('h1', "DUwDvf fontHeadlineLarge")
    print('Nama Bengkel:', nama_bengkel.text)
    
    elemen = i.find_all('div', 'Io6YTe fontBodyMedium kR99db')
    alamat=elemen[0]
    print('Alamat:', alamat.text)
    
    telepon = elemen[1]
    print('Nomor Telepon:', telepon.text)

    rating = i.find('div', 'fontDisplayLarge')
    print('Rating:', rating.text)
    
    ulasan= i.find('button',"HHrUdb fontTitleSmall rqjGif")
    print('Banyak Orang yang mengulas:',ulasan.text)
```

```python
nama_pelanggan=soup3.findAll('div', 'd4r55')
for i in nama_pelanggan:
    Nama_Pelanggan.append(i.text)

waktu_review=soup3.findAll('span','rsqaWe')
for i in waktu_review:
    Waktu_Review.append(i.text)
    
komentar=soup3.find_all('span','wiI7pd')
for i in komentar:
    Komentar.append(i.text)
    
rating_penilaian=soup3.find_all('span','kvMYJc')
for i in rating_penilaian:
    rating=i.get('aria-label')
    Rating_Penilaian.append(rating)
```

```python
data_review=pd.DataFrame(zip(Nama_Pelanggan, Waktu_Review, Rating_Penilaian, Komentar), columns=['Nama Pelanggan', 'Waktu Review','Rating Penilaian', 'Komentar'])
data_review
```
